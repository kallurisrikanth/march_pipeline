```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
from pyspark.sql import SparkSession
from pyspark.sql.functions import to_date, to_timestamp, col
```

```python
# Cell 1 - Read from bronze
df2 = spark.read.table("bronze_in")
df2.show(5)
```

```python
from pyspark.sql.functions import col, to_date

df2 = df2.select(
    to_date(col("Date"), "MMM dd, yyyy").alias("Date"),
    col("Time").cast("string"),
    col("Amount").cast("double"),
    col("Recipient").cast("string"),
    col("Type").cast("string"),
    col("Transaction_ID").cast("string"),
    col("UTR_Number").cast("bigint"),
    col("Paid_By_UPI").cast("string")
)

df2.show(10)
df2.write.mode("overwrite").saveAsTable("silver_in")
```
