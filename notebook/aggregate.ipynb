```python
df3 = spark.read.table("silver_in")
df3.show(5)
```

```python
from pyspark.sql.functions import month, col, round, sum

df4 = df3.groupBy(month('Date').alias("month")) \
         .agg(round(sum("Amount"), 2).alias("Total_Amount"))

df4.orderBy('month', ascending=False).show()

df4.write.mode("overwrite").saveAsTable("agg_sales")
```
